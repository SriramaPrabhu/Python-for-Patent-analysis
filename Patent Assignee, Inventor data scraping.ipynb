```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
import time

#PHASE-1: Downloading the latest results using key words to an excel
#STEP-1: Opening chrome
url='https://patents.google.com/advanced'
driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)

#STEP-2: Placing the input

sb=driver.find_element(By.ID,'searchInput')
sb.send_keys('("carbon capture" AND "biochar")')
time.sleep(2)


#STEP-3: Clicking Search
searchicon=driver.find_element(By.ID,'searchButton')
searchicon.send_keys(Keys.ENTER)
time.sleep(2)


#STEP-4: changing relevance to newest
relevance=driver.find_element(By.XPATH,'/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[1]/div[3]/div[2]/span/span[1]/dropdown-menu')
relevance.click()

time.sleep(2)
newest=driver.find_element(By.XPATH,'/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[1]/div[3]/div[2]/span/span[1]/dropdown-menu/iron-dropdown/div/div/div/div[2]')
newest.click()

time.sleep(2)

#STEP-5 Expanding the top companies and list
expandbutton=driver.find_element(By.XPATH,'/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[2]/div/summary-box/div[2]/div[2]/div/div[20]')
expandbutton.click()

time.sleep(2)

#STEP-6 Top companies names
n=1
basepath='/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[2]/div/summary-box/div[2]/div[2]/div/div['
Entity_list=[]
pathlist=[]
perclist=[]
while n<15:
    path=basepath+str(n)
    pathlist.append(path)
    n=n+1
    
for path in pathlist:
    Name=driver.find_element(By.XPATH,path+']/div/div/span[1]/state-modifier/a').text
    PercentPub=driver.find_element(By.XPATH,path+']/div/div/span[2]').text
    Entity_list.append([Name,PercentPub])

time.sleep(4)

#STEP-7 Top Inventor names

inventorsbutton=driver.find_element(By.XPATH,'/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[2]/div/summary-box/div[2]/div[1]/paper-tabs/div/div/paper-tab[2]')
inventorsbutton.click()

time.sleep(2)

inventor_list=[]
inv_pathlist=[]

inv_basepath='/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[2]/div/summary-box/div[2]/div[2]/div/div['
a=1
while a<15:
    inv_path=inv_basepath+str(a)
    inv_pathlist.append(inv_path)
    a=a+1
    
for inv_path in inv_pathlist:
    inventor_Name=driver.find_element(By.XPATH,inv_path+']/div/div/span[1]/state-modifier[2]/a').text
    inv_PercentPub=driver.find_element(By.XPATH,inv_path+']/div/div/span[2]').text
    inventor_list.append([inventor_Name,inv_PercentPub])

df1=pd.DataFrame(Entity_list,columns=['Name of the Entity','Percentage of Publication'])
df2=pd.DataFrame(inventor_list,columns=['Name of the Inventor','Percentage of Publication'])
df1
df2
```
